```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from astropy.io import fits
from glob import glob
```

check the nxb of Chandra and XMM-Newton near my observation time

```python
# load chandra data from Xiaoyuan results
# https://arxiv.org/pdf/2210.07284.pdf
f = open('s3_lc.txt')
lines = f.readlines()[1:]
years = np.array([line.split(' ')[0] for line in lines]).astype('float')
cts = np.array([line.split(' ')[1] for line in lines]).astype('float')
f.close()

filter = (years<8.5) & (years>6.5)
plt.plot(years[filter], cts[filter], label = 'chandra ACIS-S3 xiaoyuan mdl')
plt.axvline(7.46, c = 'k', label = 'obs time')
plt.xlabel('years - 2000')
plt.ylabel('cts')
plt.legend()
print(np.max(years))
```
